```python
import pandas as pd
import numpy as np

from sentence_transformers import SentenceTransformer, util
import tensorflow as tf
```

```python
amazon = pd.read_csv('/content/sample_data/amz_com-ecommerce_sample.csv',encoding= 'unicode_escape')
amazon_c=amazon.copy()
amazon.head()
```

```python
flipkart = pd.read_csv('/content/sample_data/flipkart_com-ecommerce_sample.csv',encoding='unicode_escape')
flipkart_c = flipkart.copy()
flipkart.head(5)
```

* Extracting only useful features


```python
amazon = amazon[['product_name','retail_price','discounted_price']]
flipkart = flipkart[['product_name','retail_price','discounted_price']]
```

**Sentence Transformers**
* Sentence transformers is a Python framework for cutting-edge sentence vector representations. Having the phrases in space allows us to calculate the distance between them, and by doing so, we can identify the ones that are most comparable in terms of their semantic meaning.

* For computing embeddings of sentences, paragraphs, and entire documents, the Sentence Transformers library is particularly effective. An embedding is just a vector representation of a text and is useful for finding how similar two texts are.

* I'll be using the "all-MiniLM-L6-v2" Sentence Transformer model for calculating semantic similarity between Amazon and Flipkart product names.

```python
# Instantiate a model of the SentenceTransformer class
model = SentenceTransformer('all-MiniLM-L6-v2')

# Compute embedding for both Amazon and Flipkart product names
amazon_prod_embs = model.encode(amazon.product_name,convert_to_tensor=True)
flipkart_prod_embs = model.encode(flipkart.product_name,convert_to_tensor=True)

# Compute cosine similarities
cosine_sim_scores = util.cos_sim(amazon_prod_embs,flipkart_prod_embs)
```

```python
# Storing the index of a similar Flipkart product corresponding to each Amazon product 
similar_prod_indexes = []

for i in range(20000):
    max_score_idxs = tf.nn.top_k(cosine_sim_scores[i],k=5,sorted=True).indices
    for idx in max_score_idxs:
        if idx != i:
            similar_prod_indexes.append(idx)
            break
```

```python
similar_prod_indexes = [x.numpy() for x in similar_prod_indexes]
```

```python
amazon.product_name[2936], flipkart.product_name[similar_prod_indexes[2936]]
```

```python
# Extracting the details such as retail price and discounted price for both Flipkart and Amazon similar products
flipkart_prod_data = []
amazon_prod_data = []

for idx, prod_name in enumerate(amazon.product_name):
    flipkart_prod_data.append(flipkart.iloc[similar_prod_indexes[idx]])
    amazon_prod_data.append(amazon.iloc[idx])
```

```python
flipkart_prod = pd.DataFrame(flipkart_prod_data)
flipkart_prod.columns = ['Product name in Flipkart','Retail Price in Flipkart','Discounted Price in Flipkart']
amazon_prod = pd.DataFrame(amazon_prod_data)
amazon_prod.columns = ['Product name in Amazon','Retail Price in Amazon','Discounted Price in Amazon']
```
